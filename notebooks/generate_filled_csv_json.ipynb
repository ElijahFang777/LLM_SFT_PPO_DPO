```python
import json
import csv
from pathlib import Path
# Input and output paths
in_path = Path("/mnt/Data2/zfang/rlhf_assignment/outputs/pairwise_compare/blind_eval.jsonl")
out_json_path = Path("/mnt/Data2/zfang/rlhf_assignment/outputs/pairwise_compare/blind_eval_filled.json")
out_csv_path = Path("/mnt/Data2/zfang/rlhf_assignment/outputs/pairwise_compare/blind_eval_filled.csv")

# Blind evaluation labels from the reviewer
labels = {
    1: "invalid",
    2: "invalid",
    3: "invalid",
    4: "b",
    5: "b",
    6: "b",
    7: "a",
    8: "a",
    9: "invalid",
    10: "invalid",
    11: "b",
    12: "invalid",
    13: "a",
    14: "b",
    15: "tie",
    16: "invalid",
    17: "tie",
    18: "a",
    19: "b",
    20: "invalid",
    21: "a",
    22: "b",
    23: "tie",
    24: "invalid",
    25: "invalid",
    26: "tie",
    27: "invalid",
    28: "a",
    29: "a",
    30: "b",
    31: "b",
    32: "tie",
    33: "tie",
    34: "invalid",
    35: "invalid",
    36: "invalid",
    37: "a",
    38: "b",
    39: "invalid",
    40: "invalid",
    41: "a",
    42: "invalid",
    43: "tie",
    44: "a",
    45: "tie",
    46: "invalid",
    47: "a",
    48: "a",
    49: "b",
    50: "tie",
    51: "invalid",
    52: "invalid",
    53: "b",
    54: "tie",
    55: "invalid",
    56: "invalid",
    57: "a",
    58: "tie",
    59: "invalid",
    60: "invalid",
}

# Read JSONL
rows = []
with open(in_path, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            row = json.loads(line)
            pair_id = int(row["pair_id"])
            row["winner"] = labels.get(pair_id, "")
            rows.append(row)

# Write JSON (array format)
with open(out_json_path, "w", encoding="utf-8") as f:
    json.dump(rows, f, ensure_ascii=False, indent=2)

# Write CSV
fieldnames = [
    "pair_id",
    "comparison",
    "prompt_id",
    "prompt",
    "response_a",
    "response_b",
    "winner",
    "notes",
]

with open(out_csv_path, "w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for row in rows:
        writer.writerow({k: row.get(k, "") for k in fieldnames})

print(f"Saved JSON: {out_json_path}")
print(f"Saved CSV:  {out_csv_path}")
```
